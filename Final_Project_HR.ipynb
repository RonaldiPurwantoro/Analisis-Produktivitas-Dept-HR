```python
def ips(x):
  if x["Age"] < 15:
    return 'Kelompok usia anak-anak'
  elif x["Age"] >= 15 and x["Age"] < 25:
    return 'Kelompok usia muda'
  elif x["Age"] >= 25 and x["Age"] < 35:
    return 'Kelompok usia pekerja awal'
  elif x["Age"] >= 35 and x["Age"] < 45:
    return 'Kelompok usia paruh baya'
  elif x["Age"] >= 45 and x["Age"] < 55:
    return 'Kelompok usia pra-pensiun'
  elif x["Age"] >= 55 and x["Age"] < 65:
    return 'Kelompok usia pensiun'
  else:
    return 'Kelompok Usia lanjut'
employee["Kelompok usia"] = employee.apply(lambda x:ips(x), axis=1)
employee
```

```python
n = len(employee_ready['Age']) # total baris
label = np.random.choice([1, 2, 3], size=(n), p=[0.7,0.2,0.1]) # membangkitkan data 0 dan 1 sebanyak n kali dengan proporsi 70%,20%,10%

# split data
df_train = employee_ready.loc[label==1,]
df_test = employee_ready.loc[label==2,]
df_validation = employee_ready.loc[label==3,]

kolom_X = ['Age', 'Female', 'Male']
kolom_y = ['Absent_Hours']

X_train, X_test, X_validation, y_train, y_test, y_validation = df_train[kolom_X], df_test[kolom_X], df_validation[kolom_X], df_train[kolom_y], df_test[kolom_y], df_validation[kolom_y]
```

```python
n = len(employee_ready['Age']) # total baris
label = np.random.choice([1, 2], size=(n), p=[0.8,0.2]) # membangkitkan data 0 dan 1 sebanyak n kali dengan proporsi 80%,20%

# split data
df_train = employee_ready.loc[label==1,]
df_test = employee_ready.loc[label==2,]

kolom_X = ['Age', 'Female', 'Male']
kolom_y = ['Absent_Hours']

X_train, X_test, y_train, y_test= df_train[kolom_X], df_test[kolom_X], df_train[kolom_y], df_test[kolom_y]
```

```python
data_regresi_train = pd.concat([X_train, y_train], axis=1)
data_regresi_train
```

```python
data_regresi_test = pd.concat([X_test, y_test], axis=1)
data_regresi_test
```

### Model Regresi Linear Berganda


```python
Input_2 = [('scale', StandardScaler()),
           ('model', LinearRegression())]
```

```python
pipe_model = Pipeline(Input_2)
pipe_model.fit(X_train, y_train)
print("Nilai R-square Train =", pipe_model.score(X_train, y_train) * 100)
```

```python
print("Nilai R-square Test =", pipe_model.score(X_test, y_test) * 100)
```

```python
print("Hasil Prediksi Y test:")
print(pipe_model.predict(X_test))
```

```python
print("nilai RMSE Test:")
print((mean_squared_error(y_test, pipe_model.predict(X_test)))**0.5)
```

### Model Regresi Linear dengan Interaksi Polinomial

```python
Input_3 = [('scale', StandardScaler()),
           ('polynomial', PolynomialFeatures(include_bias=False)),
           ('model', LinearRegression())]
```

```python
pipe_model = Pipeline(Input_3)
pipe_model.fit(X_train, y_train)
print("Nilai R-square Train =", pipe_model.score(X_train, y_train) * 100)
```

```python
print("Nilai R-square Test =", pipe_model.score(X_test, y_test) * 100)
```

```python
print("Hasil Prediksi Y test:")
print(pipe_model.predict(X_test))
```

```python
print("nilai RMSE Test:")
print((mean_squared_error(y_test, pipe_model.predict(X_test)))**0.5)
```
